###
    25. Extract a sub dataset from the original dataset. The sub dataset should contain columns 'culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g'. Perform k-means on this sub-dataset to 3 clusters. Then print the labels of each instances from the clustering


```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

df_sub = df[['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']].values

X = df_sub
y = df['label'].replace(['Adelie','Chinstrap','Gentoo'], [0,1,2])

n_clusters=3

cluster = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
y_pred = cluster.labels_
print(y_pred)
print(y.values)

centroid = cluster.cluster_centers_
centroid

inertia = cluster.inertia_  # sum of dist squire
inertia

color = ["red","blue","green"] 
fig, ax1 = plt.subplots(1)

for i in range(n_clusters):
    ax1.scatter(X[y_pred==i, 0], X[y_pred==i, 1], 
                marker='o', 
                s=8, 
                c=color[i]
               ) 
ax1.scatter(centroid[:,0], centroid[:,1],
            marker="x", 
            s=15, 
            c="black"
           ) 
plt.show()
```

###
    26. Use UMAP to to get the 2 dimensional embedding space from the original dataset
    Step:
    - Use StandardScaler from sklearn to scale the subdataset that contains features "culmen_length_mm","culmen_depth_mm","flipper_length_mm","body_mass_g"
    - Run UMAP on the scaled dataset

```python
from sklearn.preprocessing import StandardScaler
import umap
s = StandardScaler()
arr = s.fit_transform(df[['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']])
# print(arr)
reducer = umap.UMAP(random_state=21)
trans = reducer.fit_transform(arr)
# print(trans)
```
